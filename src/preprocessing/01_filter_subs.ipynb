## This notebook filters all the submissions retrieved based on the topic of interest

```python
import os
import pandas as pd
```

```python
# input data
submission_dir = ""
submission_file = os.path.join(submission_dir, "submission_data.csv")
```

```python
# output data
output_path = ""
sub_output_path = os.path.join(output_path, "submissions")
thread_output_path = os.path.join(output_path, "threads")
```

### Get all submissions matching the filtering criteria

```python
# keywords to look for
keywords = [
            "california wildfire", 
            "climate action",
            "climate change",
            "climate crisis",
            "climate emergency",
            "climate denial",
            "global warming",
            "global heating",
            "greenhouse emission",
            #"climate",
            "wildfire",
            "renewables"
        ]
```

```python
# upload submission data
sub_df = pd.read_csv(submission_file)#[['sub_id', 'sub_title']]
```

```python
# save the submission ids matching the filtering criteria
filtered_sub_ids = []
```

```python
# iterate over the sub_df rows
for index, row in sub_df.iterrows():
    # get the submission title
    sub_title = row['sub_title']
    # check whether the title contains any of the keywords
    for keyword in keywords:
        if keyword in sub_title:
            # save the submission id
            filtered_sub_ids.append(row['sub_id'])
            print("Title: added {} with k {}".format(row['sub_id'], keyword))
            break
        else: # look for the keyword in the submission selftext
            # load the submission selftext
            sub_selftext_file = os.path.join(submission_dir, "text", f"{row['sub_id']}.txt")
            with open(sub_selftext_file, "r") as f:
                sub_selftext = f.read()
            # check whether the selftext contains any of the keywords
            if keyword in sub_selftext:
                # save the submission id
                filtered_sub_ids.append(row['sub_id'])
                print("Selftext: added {} with k {}".format(row['sub_id'], keyword))
                break
```

```python
len(filtered_sub_ids)
```

### Store the filtered submissions

```python
# filter the submission data considering only 
# the submissions that contain the given keywords
filtered_sub_df = sub_df[sub_df['sub_id'].isin(filtered_sub_ids)]
filtered_sub_df.reset_index(drop=True);
```

#### Storing submission metadata

```python
# get the dir name
path_elements = submission_file.split("/") 
r_sub_output_path = os.path.join(sub_output_path, path_elements[len(path_elements)-2])

# create the output dir if it does not exist
if not os.path.exists(r_sub_output_path):
    os.makedirs(r_sub_output_path)

# get file name
f_name = submission_file.split("/")[-1]

# store the filtered submission data
filtered_sub_df.to_csv(os.path.join(r_sub_output_path, f_name), index=False)
```

#### Storing the submission text data

```python
# create the text dir
text_dir = os.path.join(r_sub_output_path, "text")

# create the text dir if it does not exist
if not os.path.exists(text_dir):
    os.makedirs(text_dir)
```

```python
# copy the text files
# corresponding to the filtered submissions
for sub_id in filtered_sub_ids:
    # get the text file
    text_file = os.path.join(submission_dir, "text", f"{sub_id}.txt")
    # copy the text file
    os.system(f"cp {text_file} {text_dir}")
```

```python
# check whether all files have been copied
len(os.listdir(text_dir)) == len(filtered_sub_ids)
```

#### Store the filtered thread text data

```python
# get the dir name
path_elements = submission_file.split("/") 
r_thread_output_path = os.path.join(thread_output_path, path_elements[len(path_elements)-2])

# create the output dir if it does not exist
if not os.path.exists(r_thread_output_path):
    os.makedirs(r_thread_output_path)
```

```python
# copy all directories associated with
# the threads of the filtered submissions
for sub_id in filtered_sub_ids:
    # get the thread dir
    _thread_dir = os.path.join(thread_dir, f"{sub_id}")
    # copy the thread dir
    os.system(f"cp -r {_thread_dir} {r_thread_output_path}")
```

```python
# check whether all files have been copied
len(os.listdir(r_thread_output_path)) == len(filtered_sub_ids)
```
